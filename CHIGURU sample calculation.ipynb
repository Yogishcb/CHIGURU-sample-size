```python
# prompt: import the file "individual_for_sample_calulation_CHIGURU" from google drive

drive.mount('/content/drive')

# Replace 'your_folder_path' with the actual path to your file in Google Drive
file_path = '/content/drive/MyDrive/Analyis/google_colab/Chiguru_sample_calculation/individual_for_sample_calulation_CHIGURU.xlsx'

try:
    df = pd.read_excel(file_path)
except FileNotFoundError:
    print(f"Error: File not found at {file_path}. Please check the path.")
except Exception as e:
    print(f"An error occurred: {e}")
else:
    # Check if necessary columns exist before proceeding
    required_cols = ['smoking_tobacco', 'past_smoking', 'smokeless_tobacco', 'alcohol_use']
    if not all(col in df.columns for col in required_cols):
      print("Error: Not all required columns are present in the dataframe.")
    else:
      # Create the 'substance_use' column
      df['substance_use'] = 'na'  # Initialize with 'na' for all rows
      for index, row in df.iterrows():
          if any(row[col] == 'yes' for col in required_cols) :
              df.loc[index, 'substance_use'] = 'using'
          elif any(pd.isna(row[col]) for col in required_cols):
              df.loc[index, 'substance_use'] = 'na'
          else:
              df.loc[index, 'substance_use'] = 'not using'
      print(df.head())
```

```python
 ## not to run this is the extra code that initially prepared

# Create 'substance_use' column
df['substance_use'] = np.where(
    (df['smoking_tobacco'] == 'yes') |
    (df['past_smoking'] == 'yes') |
    (df['smokeless_tobacco'] == 'yes') |
    (df['alcohol_use'] == 'yes'),
    'using',
    'not using'
)

# Handle missing values
df['substance_use'] = df['substance_use'].fillna('na')

display(df.head())
```
